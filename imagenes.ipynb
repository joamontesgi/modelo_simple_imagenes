```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# train
from sklearn.model_selection import train_test_split
```

```python
entrenamiento = pd.read_csv('fashion-mnist_train.csv')
pruebas = pd.read_csv('fashion-mnist_test.csv')
```

```python
X_train = entrenamiento.drop('label', axis=1)
y_train = entrenamiento['label']
X_test = pruebas.drop('label', axis=1)
y_test = pruebas['label']
```

```python
from sklearn.neural_network import MLPClassifier
import sklearn.metrics as metrics
modelo = MLPClassifier(hidden_layer_sizes=(784, 50, 50), solver='sgd', activation='logistic', max_iter=150)
modelo.fit(X_train, y_train)
y_pred = modelo.predict(X_test)
```

```python
print("Accuracy: ", metrics.accuracy_score(y_true=y_test, y_pred=y_pred), "\n")
```

```python
pesos = modelo.coefs_

# Imprimir los pesos de cada capa
for i, peso in enumerate(pesos):
    print(f"Capa {i}:")
    print(peso)
```
